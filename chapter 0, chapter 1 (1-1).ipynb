## Correlation(상관관계)

* **pearsonr(x, y):** 두 Feature에 대해서 피어슨 상관계수 값과 p-value를 도출하는 메소드


```python
features
```

## 속성(features) 중 CRIM과 ZN의 상관관계가 있는지 상관분석을 해보겠다.

```python
stats.pearsonr(features['CRIM'], features['ZN']) ### edit ###
```

```python
def corr_c(dataframe):
    col_length=len(dataframe.columns)
    idx_length=len(dataframe.index)
    for idx in range(idx_length):
        for col in range(col_length):
            dataframe.iloc[idx,col] = stats.pearsonr(features[feature_list[idx]], features[feature_list[col]])[0]
    return dataframe

def corr_p(dataframe):
    col_length=len(dataframe.columns)
    idx_length=len(dataframe.index)
    for col in range(col_length):
        for idx in range(idx_length):
            dataframe.iloc[idx,col] = stats.pearsonr(features[feature_list[idx]], features[feature_list[col]])[1]
    return dataframe
```

## features 간의 상관계수(correlation coefficient) 구하기
### Correlation Matrix : 공분산 행렬

```python
# correlation coefficient dataframe
corr_coefficient=pd.DataFrame(np.zeros((13,13))) ### edit ### #9*9 수정 중요
corr_coefficient.columns = feature_list
corr_coefficient.index = feature_list

corr_c(corr_coefficient)
```

```python
# correlation P-values dataframe
corr_p_values=pd.DataFrame(np.zeros((13,13))) ### edit ###
corr_p_values.columns = feature_list
corr_p_values.index = feature_list

corr_p(corr_p_values)
```

## features 간의 p값(p-value) 구하기

## Visualization

```python
# correlation coefficient heatmap
plt.figure(figsize=(9,9))
sns.heatmap(corr_coefficient, annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')
plt.show()
```

```python
# correlation P-values heatmap
plt.figure(figsize=(9,9))
sns.heatmap(corr_p_values, annot=True, fmt = '.2f', linewidths=.5, cmap='Reds')
plt.show()
```

```python
# scatterplots for specific variables
plt.figure(figsize=(9,9))
sns.lmplot(x="DIS", y="INDUS", data=features, line_kws={'color':"red"}, ci=None)
plt.show()
```

### 5. Interpretations and qualitative remarks based on the results
   (상관관계 분석에 따른 나의 주관적 해석)

상관계수는 두 변수가 어떤 강도로 상관이 있는지 나타내는 값이고 유의 확률(p-value)는 상관분석으로 나온 상관계수 값이 사실이 아닐(틀릴) 확률을 말하는 값이다. 
여기서 한가지 개념을 더 소개하자면 유의 수준은 내가 지정하는 기준으로 나는 0.05를 유의수준으로 설정하였다. 유의수준이 0.05라는 의미는 상관계수가 사실이 아닐 확률이 5%이고, 사실일 확률이 95%라는 의미로, 5% 정도는 틀려도 사실로 보겠다는 의미이다.
따라서, 유의수준 0.05를 기준으로 세웠을 시, 유의확률(p-value)이 0.05보다 적게 나와야 의미 있는 상관계수 값이다. 상관계수가 높아서 두 변수간에 상관관계가 높은 것처럼 보여도 유의확률(p-value)가 0.05보다 크면 둘은 상관관계가 유의미하지 않다. 
 
 귀무가설 H0 : 변수X와 변수Y가 상관 관계가 없다.
 대립가설 H1 : 변수X와 변수Y가 상관 관계가 있다.
즉, 유의 확률(p-value)가 유의수준(0.05로 설정함)보다 적어야 귀무가설을 기각 할 수 있거 상관계수 값이 인정된다는 뜻이다. -> 귀무가설을 기각한다는 뜻은 대립가설을 선택한다는 것으로 변수X와 변수Y가 상관 관계가 있다는 뜻이다.
나는 상관계수의 절댓값이 0.7이상이고 p-value가 0.05보다 작은 조건을 만족시키는 변수쌍이 상관관계가 높은 변수쌍이라고 설정하였다. 상관계수가 0.7이상이면 강한 양적 선형상관관계이고 상관계수가 –0.7이하이면 강한 음적 선형상관관계인 것이다.
Boston Housing의 변수쌍들의 강한 양적 선형관계와 강한 음적 선형관계를 분류해보면
